# ML based MALE or FEMALE Detector 

## Loading images in an data structure with tags

```python
#loading modules
import cv2
import numpy as np
import os

#preparing the labeled data
def list_all_images(image_loc,label,status):
    global train_label
    global val_label
    global val_img
    global train_img
    global height
    global width
    global depth
    
    counter=0
    
    if status == "train":
        for dir_names in os.listdir(image_loc):
            for img in os.listdir(image_loc+"/"+dir_names):
                if img.endswith(".jpg"):
                    #Resizing images
                    image =cv2.resize(cv2.imread(image_loc+"/"+dir_names+"/"+img), (height, width))
                    #Normalize
                    input_image = np.array(image)/255
                    #Changing to correct format
                    input_image = np.transpose(input_image, [2, 0, 1]).astype('float32')
                    train_img[counter]=input_image
                    train_label[counter]=label
                    counter+=1
    else:
         for img in os.listdir(image_loc):
                if img.endswith(".jpg"):
                    #Resizing
                    image =cv2.resize(cv2.imread(image_loc+"/"+img),(height,width))
                    #Normalize
                    input_image = np.array(image)/255
                    #Changing to correct format
                    input_image = np.transpose(input_image, [2, 0, 1]).astype('float32')
                    val_img[counter]=input_image
                    val_label[counter]=label
                    counter+=1

depth =3
height =28
width =28
classes =2
train_examples=123
test_examples=14

train_img = np.zeros((train_examples,depth,height,width), dtype=np.float32)
train_label = np.zeros((train_examples,classes), dtype=np.uint8)
val_img = np.zeros((test_examples,depth,height,width), dtype=np.float32)
val_label = np.zeros((test_examples,classes), dtype=np.uint8)


#Male data details
Male_file_locaiton_train= "./train/Male"
Male_file_locaiton_val="./test/Male"
male_label = 0
list_all_images(Male_file_locaiton_train,male_label,"train")
list_all_images(Male_file_locaiton_val,male_label,"val")

#Female data details
Female_file_location_trian = "./train/Female"
Female_file_location_val="./test/Female"
female_label = 1
list_all_images(Female_file_location_trian,female_label,"train")
list_all_images(Female_file_location_val,female_label,"val")

#Lets try to visualize the data
print np.shape(train_img)
print np.shape(val_img)
print np.shape(train_label)
print np.shape(val_label)

X_train = train_img
Y_train = train_label
X_val = val_img
Y_val = val_label
```

## Getting High level features in the image

```python
# importing modules
from keras.models import Sequential
from keras.layers.convolutional import Convolution2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Activation
from keras.layers.core import Flatten
from keras.layers.core import Dense
from keras.optimizers import SGD

# Sending the data through a Lenet convolutional neural to collect low level features

#Evaluation parameters
training_batch_size = 1
val_batch_size = 1
nb_epoch = 1

print "Training the DNN model\n"

#Squential layer
model = Sequential()

# 1st CONV => RELU => POOL
model.add(Convolution2D(20, 5, 5, border_mode="same",input_shape=(depth, height, width)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

# 2nd CONV => RELU => POOL
model.add(Convolution2D(50, 5, 5, border_mode="same"))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

# Fully connected layer
model.add(Flatten())
model.add(Dense(500))
model.add(Activation("relu"))

# Nodes containing the number of classes
model.add(Dense(classes))

# multinomial logistic regression - return probalities
model.add(Activation("softmax"))

#SGD Optimizer
opt = SGD(lr=0.01)
model.compile(loss="categorical_crossentropy", optimizer=opt,metrics=["accuracy"])

#Training model
print "Training the sample data \n"
model.fit(X_train, Y_train, batch_size= training_batch_size, nb_epoch= nb_epoch, verbose=2)
print "Training completed \n"

#Evaluate model
print "Evaluation of model \n"
score = model.evaluate(X_val,Y_val,batch_size=val_batch_size, verbose=0)

#Saving model weights
model.save_weights("./model_MF_10epoch.h5",overwrite=True)

print "performance of the training \n" 
print 'Test score:', score[0]
print 'Test accuracy:', score[1]
```

## Get Prediction

```python
import matplotlib.pyplot as plt

#load an image
org_img=cv2.imread("./viki.jpg")
test_img=cv2.resize(org_img,(height,width))
test_img = np.array(test_img)/255
test_img=np.transpose(test_img, [2, 0, 1]).astype('float32')
X_test = test_img[None,:,:,:]

predictions = []
predictions = model.predict(X_test)
print predictions

class_predict =  model.predict_classes(X_test)
test_probablity = model.predict_proba(X_test)
print "Class predicted is : " + str(class_predict)
print "Class probablity is : " + str(test_probablity)

%matplotlib inline
if class_predict[0] == 0:
    print "Male is detected"
    cv2.putText(org_img,"Male",(0,50), cv2.FONT_HERSHEY_SIMPLEX, 2,(255,0,0),2)
    plt.imshow(cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB))
else:
    print "Female is detected"
    cv2.putText(org_img,"Female",(0,50), cv2.FONT_HERSHEY_SIMPLEX, 2,(255,0,0),2)
    plt.imshow(cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB))
```
